```python
#before cleaning the data
from collections import defaultdict
d=defaultdict(int)
c=0
l=[]
for i in range(0, 471):
    l=(str(dataset['symptoms'][i]).lower()).strip().split(',')
    print(l)
    
```

```python
#finding count of each symptom for all the patient
z=['pain','fever','breath','respi','chest','throat','pain','musc','diarr','asymp','fati','coug','dizz','pneum']
#list of common symptoms, just the prefixes
from collections import defaultdict
d=defaultdict(int)
c=0
l=[]
for i in range(0, 471):
    l=(str(dataset['symptoms'][i]).lower()).strip().split(',')
    #print(l)
    for j in range(len(l)):
        for k in z:
            if k in l[j]:
                #print(k,l[j])
                d[k]=d[k]+1
                c=1
        if c==0:
            d[l[j]]=d[l[j]]+1
    l=[]
    c=0
```

```python
#final list of symptoms of the list
symp=[]
for i in d:
    symp.append(i)
print(symp)
```

```python
#count per symptom, one person can show multiple symptoms
print(d)
```

```python
#in percentage
for i in d:
    d[i]=(d[i]*100)/471
```
